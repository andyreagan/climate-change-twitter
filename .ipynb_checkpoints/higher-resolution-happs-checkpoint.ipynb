```python
import datetime
import matplotlib.pyplot as plt
%matplotlib inline

import re

from labMTsimple.speedy import sentiDict
LabMT = sentiDict('LabMT',datastructure='dict',stopVal=1.0)

# input the day to start and end on
startday = '2011-08-23'
endday = '2011-09-02'
startdate = datetime.datetime.strptime(startday,'%Y-%m-%d')
print(startdate.strftime('%Y-%m-%d-%H-%M'))
enddate = datetime.datetime.strptime(endday,'%Y-%m-%d')
baseresolution = datetime.timedelta(minutes=15)

def mysavefig(name,date=True):
    if date:
        plt.savefig('{0}-{1}'.format(datetime.datetime.strftime(datetime.datetime.now(),'%Y-%m-%d-%H-%M'),name))
    else:
        plt.savefig('{1}'.format(name))
```

```python
# put the resolution at which to make the timeseries
# if it's more than 15 minutes, grab more files
resolution = datetime.timedelta(minutes=15)
date = startdate
timeseries = []
time = []
while date<enddate+datetime.timedelta(days=1):
    # print(date.strftime('%Y-%m-%d-%H-%M'))
    rawtext = ''
    tmpdate = date
    while tmpdate<date+resolution:
        f = open(tmpdate.strftime('data/%Y-%m-%d-%H-%M.txt'),'r')
        rawtext += f.read()+' '
        f.close()
        tmpdate+=baseresolution
    
    rawtextdict = dict()
    words = [x.lower() for x in re.findall(r"[\w\@\#\'\&\]\*\-\/\[\=\;]+",rawtext,flags=re.UNICODE)]
    for word in words:
        if word in rawtextdict:
            rawtextdict[word] += 1
        else:
            rawtextdict[word] = 1
    score = LabMT.scoreTrie(rawtextdict)
    if score > 0:
        timeseries.append(score)
        time.append(date)
    date+=resolution
    
plt.plot(time,timeseries)
plt.ylim([4.5,7])
plt.title('15 minute resolution')
plt.ylabel('happs')
plt.xlabel('time')
```

```python
def gettimeseries(res):
    date = startdate
    timeseries = []
    time = []
    while date<enddate+datetime.timedelta(days=1):
        # print(date.strftime('%Y-%m-%d-%H-%M'))
        rawtext = ''
        tmpdate = date
        while tmpdate<date+res:
            f = open(tmpdate.strftime('data/%Y-%m-%d-%H-%M.txt'),'r')
            rawtext += f.read()+' '
            f.close()
            tmpdate+=baseresolution
        
        rawtextdict = dict()
        words = [x.lower() for x in re.findall(r"[\w\@\#\'\&\]\*\-\/\[\=\;]+",rawtext,flags=re.UNICODE)]
        for word in words:
            if word in rawtextdict:
                rawtextdict[word] += 1
            else:
                rawtextdict[word] = 1
        score = LabMT.scoreTrie(rawtextdict)
        if score > 0:
            timeseries.append(score)
            time.append(date)
        date+=res
    return time,timeseries

resolution = datetime.timedelta(hours=12)
t,h = gettimeseries(resolution)
plt.plot(t,h)
plt.ylim([4.5,7])
plt.title('12 hour resolution')
plt.ylabel('happs')
plt.xlabel('time')
```

```python
resolution = datetime.timedelta(hours=24)
t,h = gettimeseries(resolution)
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0.2,0.2,0.7,0.7])
plt.plot(t,h)
plt.ylim([5.4,6])
plt.title('1 day resolution')
plt.ylabel('happs')
plt.xlabel('time')
import matplotlib.dates as mdates
myFmt = mdates.DateFormatter('%d')
ax.xaxis.set_major_formatter(myFmt)
```

```python
titles = ('15 minutes','30 minutes','1 hour','3 hours','12 hours','1 day')
resolutions = (datetime.timedelta(minutes=15),datetime.timedelta(minutes=30),datetime.timedelta(hours=1),datetime.timedelta(hours=3),datetime.timedelta(hours=12),datetime.timedelta(hours=24))
plt.figure(figsize=(30,10))
myFmt = mdates.DateFormatter('%d')

for i,title,resolution in zip(range(len(titles)),titles,resolutions):
    t,h = gettimeseries(resolution)
    ax = plt.subplot(2,3,i+1)
    plt.plot(t,h)
    ax.xaxis.set_major_formatter(myFmt)
    # plt.ylim([5.4,6])
    plt.title(title)
    plt.ylabel('happs')
    plt.xlabel('time')
    
mysavefig('irene-higher-resolutions.pdf')
```
